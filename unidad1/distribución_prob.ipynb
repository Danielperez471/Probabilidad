----problema:
a Jules le interesa el clima. lo rastreo durante 10 dias en la mañana, a la hora de la comida y en la tarde, registrando si está soleado o nublado.
sus datos se muestran por "S" representando un clima soleado y "N" representando un clima nublado.  
1.- SSS; 2.-SSS 3.-SNN 4.-SNS 5.-NNS 6.-SSS 7.-SSN 8.-NNS 9.-SSS 10.-NSS.

Hacer una grafica de frecuencia relativa que muestre la proporcion para cada numero posible de momentos soleados en un dia para los 10 dias dados.
```
# Tiene formato de código
```



````python
import matplotlib.pyplot as plt
datos = ["SSS", "SSS", "SNN", "SNS", "NNS", "SSS", "SSN", "NNS", "SSS", "NSS"]
# Lista para almacenar la frecuencia relativa de cada cantidad posible de momentos soleados
frecuencia_relativa= [0] * 4  # Para 0, 1, 2, 3 momentos soleados
````

````python
# Contar el número de momentos soleados en cada día y actualizar la frecuencia relativa
for dia in datos:
    momentos_soleados = dia.count("S")
    frecuencia_relativa[momentos_soleados] += 1
````

````python
# Calcular la frecuencia relativa
total_dias = len(datos)
frecuencia_relativa = [frecuencia / total_dias for frecuencia in frecuencia_relativa]
````

````python
# Graficar la frecuencia relativa
momentos_posibles = [0, 1, 2, 3]
plt.bar(momentos_posibles, frecuencia_relativa, color='skyblue')
plt.title('Frecuencia relativa de momentos soleados en un día')
plt.xlabel('Número de momentos soleados')
plt.ylabel('Frecuencia relativa')
plt.xticks(momentos_posibles)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````

----problema:
Pooja tiene 5 lugares diferentes de su jardin en los que tiene 3 plantas diferentes. Cuando termina el invierno , registra si cada planta sobrevivio o murio.

Sus datos se muestran "D" para una planta que murio y "S" representa una planta que sobrevivio.

1.- DSD; 2.-DSS 3.-SSS 4.-DSD 5.-SDS

hacer una grafica de frecuencia relativa que muestre la proporcion para cada numero de plantas sobrevivientes en cada ubicacion.

````python
# Datos de las plantas sobrevivientes en cada ubicación
datos = ["DSD", "DSS", "SSS", "DSD", "SDS"]
````

````python
# Inicializar un diccionario para almacenar el conteo de plantas sobrevivientes en cada ubicación
# Las claves del diccionario representan el número de plantas sobrevivientes y los valores se inicializan en 0.
conteo_sobrevivientes = {
    0: 0,
    1: 0,
    2: 0,
    3: 0
}
````

````python
# Contar el número de plantas sobrevivientes en cada ubicación
for registro in datos:
    num_sobrevivientes = registro.count("S")
    conteo_sobrevivientes[num_sobrevivientes] += 1
````

````python
# Calcular la frecuencia relativa para cada cantidad posible de plantas sobrevivientes en cada ubicación
total_registros = len(datos)
frecuencia_relativa = {k: v / total_registros for k, v in conteo_sobrevivientes.items()}
````

````python
# Graficar la frecuencia relativa
ubicaciones = list(conteo_sobrevivientes.keys())
frecuencias = list(frecuencia_relativa.values())
````

````python
plt.bar(ubicaciones, frecuencias, color='lightgreen')
plt.title('Frecuencia relativa de plantas sobrevivientes en cada ubicación')
plt.xlabel('Número de plantas sobrevivientes')
plt.ylabel('Frecuencia relativa')
plt.xticks(ubicaciones)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````
