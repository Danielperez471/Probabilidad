```python
#CARGA Y EXPLORACION DE DATOS

import pandas as pd
import matplotlib.pyplot as plt

# Cargar el conjunto de datos
df1 = pd.read_csv('/content/grupo1.csv')

# Mostrar las primeras filas del DataFrame
print(df1.head())

# Descripción estadística de los datos
print(df1.describe())

# Visualización de los datos
# Histograma de calificaciones
plt.hist(df1['Calificacion'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Calificaciones')
plt.ylabel('Frecuencia')
plt.title('Histograma de Calificaciones')
plt.show()
```

```python
#MODELO DE REGRSION LINEAL
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr

# Preparar los datos
x = df1[['HorasEstudio']]
y = df1['Calificacion']

# Crear el modelo de regresión lineal
model = LinearRegression()
model.fit(x, y)

# Obtener la pendiente e intercepto
pendiente = model.coef_[0]
intercepto = model.intercept_
print("Ecuación de la línea de regresión:")
print(f"Calificaciones = {pendiente:.2f} * Horas de Estudio + {intercepto:.2f}")
```

```python
#VISUALIZACION DE RESULTADOS Y CALCULO DE LA CORRELACION
import numpy as np

# Graficar los datos junto con la línea de regresión
plt.scatter(df1['HorasEstudio'], df1['Calificacion'], color='blue', label='Datos')
plt.plot(df1[['HorasEstudio']], model.predict(x), color='red', label='Línea de regresión')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.title('Regresión Lineal: Horas de Estudio vs Calificaciones')
plt.legend()
plt.show()

# Calcular el coeficiente de correlación de Pearson
coef_corr, _ = pearsonr(df1['HorasEstudio'], df1['Calificacion'])
print("Coeficiente de correlación de Pearson:", coef_corr)
```
