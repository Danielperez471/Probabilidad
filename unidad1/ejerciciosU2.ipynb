###Combinaciones sin repeticion
##Ejemplo 1 nombres de personas
se trata de hacer combinaciones sin repeticion de los nombres de 10 personas
Oscar
Paco
Paty
Laura
Ruben
Lucy
Alberto
Juan
Danny
En grupos de dos

```python
import math as mt # para utilizar factorial
from itertools import combinations
from itertools import combinations_with_replacement
import numpy as np
```

```python
nombres=["Oscar","Paco", "Paty", "Laura", "Ruben", "Luis", "Lucy", "Alberto","Juan","Danny"]
n=len(nombres)
r=2 # numero por agrupacion

Cn=mt.factorial(n)/(mt.factorial(r)*mt.factorial(n-r))
print("existen ",Cn, " posibles combinaciones sin repeticion del total de ",n,
      " nombres "," en grupos de ",r," en ", r)

print ("Aqui las combinaciones")
combinaciones_nombres=list(combinations(nombres,r))

#imprime las combinaciones
for combinations in combinaciones_nombres:
  print(combinations)
```

##Ejemplo 2 caracteres alfabeticos y digitos

```python
import itertools
caracteres =["A","B","C","D", "E", "F", "0", "1", "2", "3", "4", "5"]
n=len(caracteres)
r=10 # cantidad por agrupar

Cn=mt.factorial(n)/(mt.factorial(r)*mt.factorial(n-r))
print("existen ",Cn, " posibles combinaciones sin repeticion del total de ",n,
      " caracteres "," en grupos de ",r," en ", r)

print ("Aqui las combinaciones")
combinaciones_caracteres = list(itertools.combinations(caracteres, r))

#imprime las combinaciones
for combinations in combinaciones_caracteres:
  print(combinations)
```

```python
nombres=["a","b","c","d","e","f","g"]
n=len(nombres)
r=3 #cantidad por grupos

Pr=mt.factorial(n)/(mt.factorial(n-r))
print("existen ",Pr, " posibles permutacioness sin repeticion del total de ",n,
      " nombres "," en grupos de ",r," en ", r)

print ("aqui las permutaciones")
permutaciones_nombres= list(permutations(nombres, r))
#imprime las permutaciones
for permutacion in permutaciones_nombres:
  print(permutacion)
```
