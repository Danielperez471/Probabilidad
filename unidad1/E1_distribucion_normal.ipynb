##EJERCICIO 1

una revision de los pesos de 500 alumnos del ITD, evidencio que la media es de 70kg y la desviacion estandar es de 4kg. suponiendo que los valores se distribuyen normalmente. Hallar lo siguiente:

1-generar aleatoriamente las edad de los 500 estudiantes coon una media de 70kg. y una desviacion estandar de 4kg.

2-realizar un histograma de las edades.

3-graficar la funcion de densidad de probabilidad.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

media= 70
desv_est= 4
num_datos=500

datos=np.random.normal(media, desv_est, num_datos)
datos[0:50]
x= np.linspace (media - 3 * desv_est, media + 3 * desv_est, num_datos)
#pdf = 1 /(desv_est * np.sqrt(2 * np.pi)) * np.exp(-(x - media))**2/(2 * desv_est **2)
pdf = norm.pdf(x, loc=media, scale=desv_est)

print("Imprimir los primeros 50 valores de x", x[0:50],",y los primeros 50 pdf", pdf[0:50])
plt.figure(figsize=(6,4))
plt.hist(datos, bins=15, alpha=0.4, color='b')
plt.title('Distribucion normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
plt.figure(figsize=(6,4))
plt.plot(x, pdf, label='Normal:(media:{}; ds:{})'.format(np.round(media, 2), np.round(desv_est, 2)))
plt.title('Funcion de Densidad de Probabilidad (PDF)-Distribucion normal')
plt.xlabel('Pesos en kg. de los estudiantes')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
plt.figure(figsize=(6,4))
plt.hist(datos, bins=15, alpha=0.4, density=True, color='b')
plt.plot(x,pdf, 'k', linewidth=2,)

plt.title('Distribucion normal ')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

##ESTUDIANTES QUE PESAN MENOS DE 75KG Y CUANTOS ALUMNOS DE LOS 500 PESAN MENOS DE 75KG

```python
plt.figure(figsize=(6,4))
media=70
desv_est=4
total_estudiantes=500
#variable para el calculo de la porbabilidad de que un estudiante pese menos de 75kg
peso_limite=75
#usamos la funcion de distribucion acumulativa(CDF) de la distribucion normal
prob_menos_75kg=norm.cdf(peso_limite, media, desv_est)
#calculo del numero de estudianttes que pesan menos de 75kg
estudiantes_menos_75kg= total_estudiantes* prob_menos_75kg

print("Probabilidad de que un estudiante pese menos de 75kg: ", prob_menos_75kg)
print("Numero de estudiantes que pesan menos de 75kg: ", estudiantes_menos_75kg)
#graficar la distribucion normal
x=np.linspace(media-4*desv_est,media+4*desv_est,1000)
y=norm.pdf(x,media, desv_est)

plt.plot(x,y,'b',linewidth=2)
plt.fill_between(x,y,where=(x<=peso_limite),color='skyblue',alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 75kg.\n $\mu=70$, $\sigma=4$, x=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

##ESTUDIANTES QUE PESAN MAS DE 79KG Y CUANTOS ALUMNOS DE LOS 500 PESAN MAS DE 79KG

##ESTUDIANTES QUE PESAN ENTRE 68 Y 72KG Y CUANTOS ALUMNOS DE LOS 500 PESAN ENTRE 68 Y 72KG

```python
plt.figure(figsize=(6,4))

# Parámetros de la distribución normal
media = 70
desv_est = 4
total_estudiantes = 500
peso_limite_inferior = 68
peso_limite_superior = 72

# Usamos la función de distribución acumulativa (CDF) de la distribución normal
prob_menos_68kg = norm.cdf(peso_limite_inferior, media, desv_est)
prob_menos_72kg = norm.cdf(peso_limite_superior, media, desv_est)

# Probabilidad de estudiantes que pesan más de 68kg y menos de 72kg
prob_entre_68_y_72kg = prob_menos_72kg - prob_menos_68kg

# Número de estudiantes que pesan más de 68kg y menos de 72kg
estudiantes_entre_68_y_72kg = total_estudiantes * prob_entre_68_y_72kg

print("Probabilidad de estudiantes que pesan más de 68kg y menos de 72kg:", prob_entre_68_y_72kg)
print("Número de estudiantes que pesan más de 68kg y menos de 72kg:", estudiantes_entre_68_y_72kg)

# Graficar la distribución normal
x = np.linspace(media - 4 * desv_est, media + 4 * desv_est, total_estudiantes)
y = norm.pdf(x, media, desv_est)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=((x >= peso_limite_inferior) & (x <= peso_limite_superior)), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese entre 68kg y 72kg.\n$\mu=70$, $\sigma=4$, n=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```
