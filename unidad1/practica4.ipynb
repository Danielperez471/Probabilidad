##EJEMPLO 3: Histogramas de 2 grupos

1. en este ejemplo generamos 100 numeros aleatorios reales que representa la produccion diaria de 100 telares (medidas en metros).
2. generar otra lista con la produccion de 100 telares para una segunda fabrica.
3. con estas listas de valores se pretende determinar las medidas de tendencia central, es decir, la media, la moda, la mediana, la desviacion estandar.
4. graficar un histograma para cada fabrica.

```python
from statistics import *
```

```python
#libreria para generar numeros aleatorios
import random
```

```python
fabrica1=[]
for i in range (100):
  numero_aleatorio= round(random.uniform(50,100),2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
fabrica2=[]
for i in range (100):
  numero_aleatorio2= round(random.uniform(50,100),2)
  fabrica2.append(numero_aleatorio2)
print(fabrica2)
```

```python
#libreria de python para graficar valores
import matplotlib.pyplot as plt
```

```python
plt.hist(fabrica1)
plt.title("Produccion diaria de la fabrica 1")
```

```python
plt.hist(fabrica2)
plt.title("Produccion diaria de la fabrica 2")
```

```python
media=mean(fabrica1)
moda = mode(fabrica1)
mediana = median(fabrica1)

print ("valores de la fabrica 1")
print ("la media es: ",media)
print("la mediana es: ",mediana)
print("la moda es: ",moda)
```

```python
media=mean(fabrica2)
moda = mode(fabrica2)
mediana = median(fabrica2)

print ("valores de la fabrica 2")
print ("la media es: ",media)
print("la mediana es: ",mediana)
print("la moda es: ",moda)
```

```python
#generar 2 subplot de 1 renglon por 2 columnas para desplegar los 2 histogramas,
# una al lado de la otra
fig, (ax1, ax2)=plt.subplots(nrows=1,ncols=2)
```

```python
#generar 2 subplot de 1 renglon por 2 columnas para desplegar los 2 histogramas,
# una al lado de la otra
fig, (ax1, ax2)=plt.subplots(nrows=1,ncols=2, figsize=(10,5))
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show()
```

##Tablas de frecuencias

```python
#generamos una lista vacia que representa la edad de un grupo de personas
edades =[]
for i in range (100):
  n_aleatorio = random.randint(18,70)
  edades.append(n_aleatorio)
print (edades)
```

```python
import pandas as pd
```

```python
df_edades=pd.DataFrame(edades, columns=["Edad"])
df_edades
```

```python
#calcular la tabla de distribucion de frecuencias
#(frec. relativa, frec.porcentual, frec. acumulada)
t_frec=df_edades["Edad"].value_counts().reset_index()
t_frec.columns=["Edad","Frecuencia"]
t_frec=t_frec.sort_values(by="Edad")
t_frec["frec. relativa"]=t_frec["Frecuencia"]/100
t_frec["frec. porcentual"]= t_frec["Frecuencia"]/100*100
t_frec["frec. acumulada"]=t_frec["Frecuencia"].cumsum()
t_frec
```

##Regla de Sturges
1. es una regla que sirve para calcular el numero de clases o intervalos idoneo, en los que se debe dividir un conjunto de datos

$
c=1+log_2(N)
$

```python
#libreria de funciones matematicas
import math
```

```python
df_edades_sturges= pd.DataFrame({"Edad":edades})
df_edades_sturges
```

```python
plt.hist(edades, alpha=0.5, rwidth=0.85)
plt.title("Histograma de edades basado en valores estandar de matplotlib" )
```

```python
#calcular el numero de intervalos usando la regla de Sturges
n=len(edades)
k=int(1+math.log2(n))
print("valor de n=",n,"\nk o numero de intervalos= ",k)
```

```python
#creamos la tabla de frecuencias con los intervalos
t_frec_sturges=pd.cut(df_edades_sturges["Edad"],bins=k).value_counts().reset_index()
t_frec_sturges.columns=["clase","Frecuencia"]
t_frec_sturges=t_frec_sturges.sort_values(by="clase")
t_frec_sturges["frec. relativa"]=t_frec_sturges["Frecuencia"]/n
t_frec_sturges["frec. porcentual"]= t_frec_sturges["Frecuencia"]/n*100
t_frec_sturges["frec. acumulada"]=t_frec_sturges["Frecuencia"].cumsum()
t_frec_sturges
```

```python
plt.hist(df_edades_sturges["Edad"],bins=k, rwidth=0.85)
plt.title("Histograma de edades basado en la regla de Sturges")
```
